##Mayor numero
Realice un programa en Python que determine el mayor de tres números

##EJERCICIO EDADES

Diseñe un programa que me genere un listado con las edades de N habitantes de una ciudad, donde la cantidad de habitantes sea un valor randomico entre 1000 y 50000 y la edad de cada persona también debe ser un randomico entre 1 y 100 años. Al final del el programa me debe indicar cuanto es la población total y el porcentaje de cuantos son mayores de edad (>=18) y cuantos no.

* Metodo clasico:

```
amnt = random.randrange(1000, 50000)
elders,young = 0,0
lista = [random.randrange(1,100) for i in range(amnt)]
for i in range(0,len(lista)):
  if (lista[i]>=18): elders+=1
  elif (lista[i]<=18): young+=1
print("Total population: ", amnt)
print("Over 18 percentage: ", '{0:.3g}'.format((elders/amnt)*100), "%")
print("Under 18 percentage: ", '{0:.3g}'.format((young/amnt)*100), "%")
```

* Metodo python:

```
import random
elders=0
people = [random.randrange(1,100) for i in range(random.randrange(1000, 50000))]
for i in range(0,len(people)): 
  if people[i]>=18: elders+=1
print("Total population: ", len(people))
print("Over 18 percentage: ", '{0:.3g}'.format((elders/len(people))*100), "%")
print("Under 18 percentage: ", '{0:.3g}'.format(((len(people)-elders)/len(people))*100), "%")
```

Diseñe un programa (máximo 5 líneas) donde 6 candidat@s se disputan la alcaldía (ayuntamiento municipal), los candidat@s de los siguientes partidos:

###Numpy Gobernacion
Diseñe un programa donde 6 candidat@s se presentan a la disputa para la Gobernación del departamento (estado).
Los votos de cada candidat@s son generados de forma aleatoria entre 1 y 4500 votos máximo para cada uno de los 123 municipios del departamento, el software debe determinar quién fue ganador@.
Nota: Creamos una matriz llamada votos, compuesta por 123 filas (municipios) y 6 columnas (candidat@s) y cuyo valor de
cada celda sea un randomico entre 1 y 3000

```
import numpy as np; from random import randint
votos_2021 = np.random.randint(3000,size=(123,6))
tot_votos = np.array([[candidato+1,votos_2021[:,candidato].sum()] for candidato in range(len(votos_2021[0]))])
ganador = np.where(tot_votos == tot_votos[:,1].max())
print("El candidato ganador fue el: "+str(tot_votos[ganador[0]])+"\n El total de votos por candidato es \n"+str(tot_votos))
```

###Numpy F1 Race

Crear un programa en PYTHON para simular una carrera de formula uno, en donde existen 5 vehículos: Renault(1), Ferrari(2), Mercedes(3), Ford(4) y Red Bull(5).

La carrera esta diseñada para 4 vueltas donde el tiempo mínimo es 8.2 segundos y el máximo es 10
segundos.

Diseñe la matriz de 5 x 4 donde determine cual es el vehículo ganador donde los tiempos de cada vehículo por
vuelta se un randomico.



```
import numpy as np
matriz = np.random.uniform(8.2,10,size=(4,5))
tot_vueltas = np.array([[corredor+1,matriz[:,corredor].sum()] for corredor in range(len(matriz[0]))])
ganador=np.where(matriz == matriz[:,1].min())
print ("Race results:\n"+str(matriz)+"\nThe winner is: "+str(tot_vueltas[ganador[0]]))
```
